```python
import numpy as np
```

```python
class test_gradient():
    def __init__(self,k,alpha):
        np.random.seed(1)
        self.k=k # nombre de bras
        self.alpha=alpha # hyperparametre
        self.H = np.zeros((self.k)) # preference
        self.mu= np.random.normal(size=self.k) # récompense optimale
        
    def display(self):
        print(f"k={self.k}")
        print(f"alpha={self.alpha}")
        print(f"H(a)={self.H}")
        print(f"mu(a)={self.mu}")
        
    def one_step(self):
        # calcul les probabilités
        exp=np.exp(self.H)
        p=np.zeros((self.k))
        # votre code
        return(p)
    
    def env(self,arm):
        # renvoie la récompense
        q=np.random.normal(self.mu[arm],0.25)
        return(q)
    
    def simu(self,nbstep):
        # Simulation
        AverageReward=5 # baseline
        self.H = np.zeros((self.k)) # préférence
        indices=np.arange(self.k)
        for i in range(nbstep):
            p=self.one_step()
            # votre code
            # selection du bras en fonction des probabilités
            # calcul de la réécompense
            # mise à jour préférence du bras choisi
            # mise à jour préférence des autres bras
            arm=np.random.choice(indices,p=p)
            reward=self.env(arm)
            self.H[arm] = self.H[arm] +self.alpha*(reward-AverageReward)*(1-p[arm])
            
            for j in range(self.k):
                if j != arm:
                    self.H[j]=self.H[j]-self.alpha*(reward-AverageReward)*p[j]
            AverageReward=(AverageReward*i+reward)/(i+1)
            print(arm,reward,AverageReward)
```

```python
s=test_gradient(10,0.1)
```

```python
s.display()
```
